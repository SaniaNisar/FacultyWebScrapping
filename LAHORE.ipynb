<h1 style="font-size:3rem; color: red;"> Web Scrapping</h1>

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
URL = "http://lhr.nu.edu.pk/faculty/"
r = requests.get(URL)
print(r)
```

```python
soup = BeautifulSoup(r.content, 'html.parser')
```

```python
print(soup.prettify())
```

```python
faculty = soup.find_all('div', class_='facultyCard')
```

```python
print(faculty)
```

```python
ID = []                    # Integer
Name = []                  # String
Designation = []           # String
HEC_Approved_PHD_Supervisor = []  # Bool
Highest_Education = []     # String
Email = []                 # EMAIL
Department = []            # String
Extension = []             # Integer
ImageURL = []              # URL
```

```python
for mem in faculty:
    try:
        ID.append(int(mem.find('a',class_='faculty-link')['href'].split('/')[-1]))
        
    except:
        ID.append(-1)
        
    Name.append(mem.find('h5', class_='text-center').text.strip())

    Designation.append(mem.find('p', class_='small').text.strip().split('\n')[0])

    Email.append(mem.find('p', class_='mb-0').text.strip())

    
    try:
         hec_status = mem.find('p', class_='small').text.strip().split('\n')[2]
         HEC_Approved_PHD_Supervisor.append(True if hec_status else False)
    except IndexError:
        HEC_Approved_PHD_Supervisor.append(False)
        
    ImageURL.append("https://lhr.nu.edu.pk" + mem.find('img', class_="card-img-top")['src'])
```

```python
print(ID,
Name,
Designation,
HEC_Approved_PHD_Supervisor,
Email,
ImageURL)
```

```python
print(len(ID))
```

```python
department_mapping = {
    'fsc': 'Fast School of Computing',
    'ee': 'Electrical Engineering',
    'cv': 'Civil Engineering',
    'fsm': 'Fast School of Management',
    'ss': 'Sciences and Humanities'
}
```

```python
for dept_id, dept_name in department_mapping.items():
    department_div = soup.find('div', id=dept_id)
    if department_div:
        department_faculty = department_div.descendants
        for element in department_faculty:
            if element.name == 'h5' and 'text-center' in element.get('class', []):
                Department.append(dept_name)
```

```python
print(Department)
```

```python
for mem in faculty:
    memURL = "https://lhr.nu.edu.pk" + mem.find('a',class_='faculty-link')['href']
    memPage = requests.get(memURL)
    memContent = BeautifulSoup(memPage.content, 'html.parser')
    memProfile = memContent.find('div',class_='container')
    try:
        Highest_Education.append(memProfile.find('li').text)
    except:
        Highest_Education.append('N/A')
          
    try:
        Extension.append(int(memProfile.find('span',class_="small").text.split()[-1][-3:]))
    except:
        Extension.append(-1)
```

```python
print(len(Highest_Education))
```

```python
faculty_df = pd.DataFrame({'ID': ID , 'Name': Name , 'Designation': Designation , 'HEC_Approved_PHD_Supervisor': HEC_Approved_PHD_Supervisor ,
                       'Highest_Education': Highest_Education , 'Email' : Email , 'Department': Department , 'Extension': Extension ,
                       'ImageURL': ImageURL })
```

```python
faculty_df.to_csv('lhr.csv')
```
