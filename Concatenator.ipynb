```python
import pandas as pd
```

```python
# Load each CSV file manually
df1 = pd.read_csv('lhr.csv')
df2 = pd.read_csv('isl.csv')
df3 = pd.read_csv('khi.csv')
df5 = pd.read_csv('psh.csv')
df4 = pd.read_csv('chin.csv')
```

```python
combined_df = pd.concat([df1, df2, df3, df5, df4], ignore_index=True)
```

```python
combined_df.head()
```

```python
combined_df.to_csv('combined_file.csv', index=False)
```

```python
faculty = pd.read_csv('combined_file.csv')
```

```python
lastDigit = 5
```

```python
fracVal = lastDigit / 10 if lastDigit != 0 else 0.1
```

```python
print(fracVal)
```

```python
sample_ = faculty.sample(frac=fracVal, random_state=1)
```

```python
sample_.to_csv('sample.csv', index=False)
```
