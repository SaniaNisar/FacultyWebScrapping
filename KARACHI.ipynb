```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
URL = "https://khi.nu.edu.pk/faculty-php/"
r = requests.get(URL)
print(r)
```

```python
soup = BeautifulSoup(r.content, 'html.parser')
```

```python
print(soup.prettify())
```

```python
faculty = soup.find_all('div', class_='gdlr-core-personnel-list')
print(len(faculty))
```

```python
ID = []                    # Integer
Name = []                  # String
Designation = []           # String
HEC_Approved_PHD_Supervisor = []  # Bool
Highest_Education = []     # String
Email = []                 # EMAIL
Department = []            # String
Extension = []             # Integer
ImageURL = []              # URL
```

```python
for mem in faculty:
    try:
        Name.append(mem.find('h3', class_='gdlr-core-personnel-list-title').text.split(',')[0])
    except:
        Name.append(mem.find('h3', class_='gdlr-core-personnel-list-title').text)

    try:
        ImageURL.append(mem.find('img')['src'])
    except TypeError:
        ImageURL.append("N/A")

    Designation.append(mem.find('div', class_='gdlr-core-personnel-list-position').text)
    Email.append(mem.find('div', class_='kingster-type-email').text)
    Department.append("FAST School of Computing")
    
    profile_link = mem.find('a', class_='gdlr-core-personnel-list-button')['href']
    profile = requests.get(profile_link)
    profile_content = BeautifulSoup(profile.content, 'html.parser')
    ID.append(int(profile_content.find('body').get('class')[3][-4:]))
    HEC = profile_content.find('span', class_='gdlr-core-title-item-caption').text.strip().split('\n')

    if len(HEC) == 3:
        HEC_Approved_PHD_Supervisor.append(True)
    else:
        HEC_Approved_PHD_Supervisor.append(False)

    try:
        Extension.append(int(mem.find('div', class_='kingster-type-phone').text))
        
    except:
        Extension.append(-1)

    profile_info = profile_content.findAll('span',class_={'gdlr-core-title-item-caption' : True , "gdlr-core-icon-list-content" : True })
    check = False

    try:
        for p in profile_info:
            text = p.text.split()[0]
            if text == 'PhD' and check==False:
                Highest_Education.append(p.text)
                check = True
            elif text == 'PostDoc' and check==False:
                Highest_Education.append(p.text)
                check = True
            elif text == 'MS' and check==False:
                Highest_Education.append(p.text)
                check = True
            elif text == 'ME' and check==False:
                Highest_Education.append(p.text)
                check = True
            elif text == 'MCS,' and check==False:
                Highest_Education.append(p.text)
                check = True
            elif text == 'MPhil' and check==False:
                Highest_Education.append(p.text)
                check = True
            elif text == 'BS' and check==False:
                Highest_Education.append(p.text)
                check = True
            elif text == 'BE' and check==False:
                Highest_Education.append(p.text)
                check = True
        if check==False:
            Highest_Education.append('N/A')
    except:
        Highest_Education.append('N/A')
```
